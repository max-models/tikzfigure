```python
from maxtikzlib.figure import TikzFigure
```

```python
# source: https://github.com/ltrujello/Tikz-Python/blob/main/examples/lorenz/lorenz.py
import numpy as np
from scipy.integrate import odeint

#tikz.set_tdplotsetmaincoords(60, 45)
#tikz.options = "tdplot_main_coords"

# lorenz parameters
rho = 28.0
sigma = 10.0
beta = 8.0 / 3.0


# Next state according to the ODEs
def next(*state):
    x, y, z = state[0][0], state[0][1], state[0][2]
    return sigma * (y - x), x * (rho - z) - y, x * y - beta * z


# Set initial conditions and time steps
initial = [0.1, 0.1, 0.2]
t = np.arange(
        0.0, 20.0, 0.01
    )

# Solve for the next positions, scale them
states = odeint(func=next, y0=initial, t=t)
states = states * 0.25

# We convert points from np.array to tuple for Tikz
tuple_states = []
for state in states:
    tuple_states.append((state[0], state[2]))

# Plot the lorenz system
fig = TikzFigure()
path_actions = ["->", "draw", "line width=1", "color=black"]
fig.add_path(tuple_states, path_actions=path_actions, layer=0)
fig.add_node(*tuple_states[0], content=f'Initial: {tuple(initial)}')
fig.show()
```
