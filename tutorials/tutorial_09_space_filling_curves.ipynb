# Tutorial 9 (Advanced) — Space-Filling Curves

This showcase tutorial demonstrates tikzpics' layer system and coordinate path drawing
through two classic **space-filling curves**:

- **Z-order curve** (Morton code) — visits grid cells in a Z-shaped pattern at each
  level of recursion.
- **Hilbert curve** — visits every cell while preserving spatial locality.

The background grid is rendered on low-numbered layers (yellow → blue → red for
three recursive levels); the curve path is drawn on top on layer 5.

Key techniques used:
- helper functions that call `fig.draw()` many times
- layered filled squares to show hierarchical structure
- `center=True` on paths for TikZ's `triangle` arrow tip
- `figure_setup` to configure the default arrow style

```python
from tikzpics import TikzFigure
```

```python
def add_grid(tikz, x0, y0, L, N, options, layer=1):
    """Draw an N×N grid of lines inside the square [x0, x0+L] × [y0, y0+L]."""
    x1, y1, delta = x0 + L, y0 + L, L / N
    for i in range(N + 1):
        y = y0 + i * delta
        tikz.draw([[x0, y], [x1, y]], options=options, layer=layer)
        x = x0 + i * delta
        tikz.draw([[x, y0], [x, y1]], options=options, layer=layer)


def add_square(tikz, x0, y0, L, options, layer=1):
    """Draw (or fill) a single square."""
    pts = [[x0, y0], [x0 + L, y0], [x0 + L, y0 + L], [x0, y0 + L], [x0, y0]]
    tikz.draw(pts, options=options, layer=layer)


def draw_background(tikz):
    """Three-level hierarchical background used by both curve examples."""
    # Filled backgrounds (low layers)
    add_square(tikz, 0, 0, 8, ["draw", "line width=0", "fill=yellow!10!white"], layer=0)
    add_square(tikz, 4, 2, 4, ["draw", "line width=0", "fill=blue!10!white"], layer=1)
    add_square(tikz, 5, 3, 1, ["draw", "line width=0", "fill=red!10!white"], layer=2)

    # Grid lines (higher layers)
    add_grid(tikz, 0, 0, 8, 4, ["draw", "line width=2", "color=black"], layer=3)
    add_grid(tikz, 4, 2, 4, 4, ["draw", "line width=2", "color=blue"], layer=4)
    add_grid(tikz, 5, 3, 1, 2, ["draw", "line width=2", "color=red"], layer=5)
```

## Z-order curve

At each recursion level the four quadrants are visited in a Z pattern:
top-left → top-right → bottom-left → bottom-right.

```python
tikz = TikzFigure(figure_setup=">={Triangle[scale=1]}")
draw_background(tikz)

z_path = [
    [1, 7],
    [3, 7],
    [1, 5],
    [3, 5],
    [5, 7],
    [7, 7],
    [4.5, 5.5],
    [5.5, 5.5],
    [4.5, 4.5],
    [5.5, 4.5],
    [6.5, 5.5],
    [7.5, 5.5],
    [6.5, 4.5],
    [7.5, 4.5],
    [1, 3],
    [3, 3],
    [1, 1],
    [3, 1],
    [4.5, 3.5],
    [5.25, 3.75],
    [5.75, 3.75],
    [5.25, 3.25],
    [5.75, 3.25],
    [4.5, 2.5],
    [5.5, 2.5],
    [6.5, 3.5],
    [7.5, 3.5],
    [6.5, 2.5],
    [7.5, 2.5],
    [5, 1],
    [7, 1],
]

for i in range(len(z_path) - 1):
    tikz.draw(
        [z_path[i], z_path[i + 1]],
        options=["->", "draw", "line width=1", "color=black"],
        layer=6,
        center=True,
    )

tikz.show()
```

## Hilbert curve

The Hilbert curve visits every cell in the grid while ensuring that spatially
adjacent cells are also adjacent in the traversal order — making it valuable
for locality-preserving mappings.

```python
tikz = TikzFigure(figure_setup=">={Triangle[scale=1]}")
draw_background(tikz)

hilbert_path = [
    [1, 1],
    [3, 1],
    [3, 3],
    [1, 3],
    [1, 7],
    [3, 7],
    [3, 5],
    [4.5, 4.5],
    [5.5, 4.5],
    [5.5, 5.5],
    [4.5, 5.5],
    [5, 7],
    [7, 7],
    [7.5, 5.5],
    [6.5, 5.5],
    [6.5, 4.5],
    [7.5, 4.5],
    [7.5, 2.5],
    [6.5, 2.5],
    [6.5, 3.5],
    [5.75, 3.75],
    [5.75, 3.25],
    [5.25, 3.25],
    [5.25, 3.75],
    [4.5, 3.5],
    [4.5, 2.5],
    [5.5, 2.5],
    [5, 1],
    [7, 1],
]

for i in range(len(hilbert_path) - 1):
    tikz.draw(
        [hilbert_path[i], hilbert_path[i + 1]],
        options=["->", "draw", "line width=1", "color=black"],
        layer=6,
        center=True,
    )

tikz.show()
```
