```python
from maxtikzlib.figure import TikzFigure
import numpy as np
```

```python
fig = TikzFigure()

n1 = fig.add_node(0, 0, shape="circle", color="white", fill="blue", content="Hello!")
n2 = fig.add_node(5, 0, shape="circle", color="white", fill="red", content="Hi!")

fig.add_path([n1, n2], path_actions=["->", "line width=2"], color="gray")

fig.show()
```

```python
fig = TikzFigure()

nodes_heart = [
    [2, 0],
    [0, 3.5],
    [1, 4.5],
    [2, 3.0],
    [3, 4.5],
    [4, 3.5],
    [3, 2],
]


# TODO: Not every corner should be rounded!
fig.add_path(
    nodes_heart,
    path_actions=["draw", "rounded corners=20pt", "line width=4"],
    layer=0,
    cycle=True,
    color="red",
    fill="red!40!white",
    comment="Heart shape",
)

fig.show()
```

```python
fig = TikzFigure()
nodes_star = [
    [0, 0],
    [1, 2.5],
    [2, 0],
    [-0.5, 1.5],
    [2.5, 1.5],
    [0, 0],
]

fig.add_path(
    nodes_star,
    path_actions=["draw", "line width=3"],
    layer=0,
    cycle=True,
    color="gold",
    fill="yellow!50!white",
    comment="Star shape",
)
fig.show()
```

```python
fig = TikzFigure()

N = 200
theta = np.linspace(0, 8 * np.pi, N)
r = np.linspace(0.1, 2, N)
x_spiral = r * np.cos(theta) + 5
y_spiral = r * np.sin(theta) + 2
nodes_spiral = list(zip(x_spiral, y_spiral))

fig.add_path(
    nodes_spiral,
    path_actions=["draw", "line width=2"],
    layer=1,
    color="teal",
    comment="Spiral curve",
)
fig.show()
```

```python
fig = TikzFigure()


# Circle (sun core)
theta = np.linspace(0, 2 * np.pi, 50)
x_circle = np.cos(theta) * 1 + 6
y_circle = np.sin(theta) * 1 + 6
nodes_circle = list(zip(x_circle, y_circle))

fig.add_path(
    nodes_circle,
    path_actions=["draw", "line width=3"],
    layer=0,
    cycle=True,
    color="orange",
    fill="yellow!50!white",
    comment="Sun core",
)

# Rays
for angle in np.linspace(0, 2 * np.pi, 12, endpoint=False):
    x0, y0 = 6 + np.cos(angle), 6 + np.sin(angle)
    x1, y1 = 6 + 1.5 * np.cos(angle), 6 + 1.5 * np.sin(angle)
    fig.add_path(
        [(x0, y0), (x1, y1)],
        path_actions=["draw", "line width=2"],
        layer=0,
        color="orange",
        comment="Sun ray",
    )

# TODO: Draw the rays with a tikz loop!
# num_rays = 12
# with fig.loop("k", range(num_rays), comment="Sun rays") as loop_k:
#    angle = 2*np.pi/num_rays * loop_k.index  # angle step
#    x0, y0 = 6 + np.cos(angle), 6 + np.sin(angle)
#    x1, y1 = 6 + 1.5*np.cos(angle), 6 + 1.5*np.sin(angle)
#    loop_k.add_path(
#       [(x0, y0), (x1, y1)],
#        path_actions=["draw", "line width=2"],
#       layer=0,
#        color="orange",
#    )

fig.show()
```
