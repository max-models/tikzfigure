## Basic usage

`fig.midpoint(node1, node2)` places a new node exactly halfway between two existing nodes.
You can pass either `Node` objects or node label strings.

```python
fig = TikzFigure()

n1 = fig.add_node(
    0, 0, label="A", content="A", shape="circle", fill="blue!30", draw="blue"
)
n2 = fig.add_node(
    4, 0, label="B", content="B", shape="circle", fill="red!30", draw="red"
)

# Midpoint using Node objects
mid = fig.midpoint(
    n1, "B", label="M", content="M", shape="circle", fill="green!30", draw="green"
)

fig.draw([n1, n2], options=["dashed", "gray"])

fig.show()
print(f"Midpoint coordinates: ({mid.x}, {mid.y})")
```

## Using label strings

You can also refer to nodes by their label strings instead of keeping a reference to the object.

```python
fig = TikzFigure()

fig.add_node(
    0, 2, label="P", content="P", shape="circle", fill="orange!40", draw="orange"
)
fig.add_node(
    6, 2, label="Q", content="Q", shape="circle", fill="orange!40", draw="orange"
)

# Midpoint using label strings
mid = fig.midpoint(
    "P", "Q", label="PQ", content="mid", shape="circle", fill="white", draw="black"
)

fig.draw(["P", "Q"], options=["dashed", "gray"])

fig.show()
```

## Edge labels on a graph

A common use case is placing a label node at the midpoint of an edge.

```python
fig = TikzFigure(figure_setup="every node/.append style={draw, minimum size=0.7cm}")

# Triangle vertices
fig.add_node(0, 0, label="A", content="A", shape="circle", fill="blue!20")
fig.add_node(4, 0, label="B", content="B", shape="circle", fill="blue!20")
fig.add_node(2, 3, label="C", content="C", shape="circle", fill="blue!20")

# Edges
fig.draw(["A", "B"], color="black")
fig.draw(["B", "C"], color="black")
fig.draw(["A", "C"], color="black")

# Midpoint nodes on each edge as edge labels
fig.midpoint("A", "B", label="AB", content="$c$", draw="none", fill="yellow!60")
fig.midpoint("B", "C", label="BC", content="$a$", draw="none", fill="yellow!60")
fig.midpoint("A", "C", label="AC", content="$b$", draw="none", fill="yellow!60")

fig.show()
```

## Recursive midpoints — binary subdivision

By repeatedly calling `midpoint` you can bisect a segment multiple times.

```python
fig = TikzFigure()

node_style = dict(
    shape="circle", minimum_size="6pt", fill="black", draw="black", inner_sep="0pt"
)

left = fig.add_node(0, 0, label="L", content="", **node_style)
right = fig.add_node(8, 0, label="R", content="", **node_style)

fig.draw([left, right], color="gray")

# First-level midpoint
m1 = fig.midpoint(left, right, label="M1", content="", **node_style)

# Second-level midpoints
m2a = fig.midpoint(left, m1, label="M2a", content="", **node_style)
m2b = fig.midpoint(m1, right, label="M2b", content="", **node_style)

# Third-level midpoints
fig.midpoint(left, m2a, label="M3a", content="", **node_style)
fig.midpoint(m2a, m1, label="M3b", content="", **node_style)
fig.midpoint(m1, m2b, label="M3c", content="", **node_style)
fig.midpoint(m2b, right, label="M3d", content="", **node_style)

fig.show()
```

## Centroid of a triangle via iterated midpoints

The centroid of a triangle lies at the midpoint between any vertex and the midpoint of the opposite edge.
We can compute it with two `midpoint` calls.

```python
fig = TikzFigure()

# Triangle vertices
A = fig.add_node(
    0, 0, label="A", content="$A$", shape="circle", fill="blue!20", draw="blue"
)
B = fig.add_node(
    6, 0, label="B", content="$B$", shape="circle", fill="blue!20", draw="blue"
)
C = fig.add_node(
    2, 4, label="C", content="$C$", shape="circle", fill="blue!20", draw="blue"
)

# Triangle edges
fig.draw([A, B, C], cycle=True, color="blue")

# Midpoint of the opposite edge to A (i.e. midpoint of BC)
mid_BC = fig.midpoint(
    B,
    C,
    label="MBC",
    content="",
    shape="circle",
    minimum_size="4pt",
    fill="gray",
    draw="gray",
    inner_sep="0pt",
)

# Median from A to mid_BC
fig.draw([A, mid_BC], options=["dashed"], color="gray")

# Centroid = midpoint of A and mid_BC, shifted by the remaining third
# More directly: centroid x = (Ax + Bx + Cx) / 3 — but using midpoints:
# midpoint(A, midpoint(B, C)) gives the point 1/4 of the way; the centroid
# is 2/3 along the median. We can get it as midpoint(midpoint(A, B), C).
centroid = fig.midpoint(
    fig.midpoint(A, B, label="_MAB", content=""),  # midpoint AB
    C,
    label="G",
    content="$G$",
    shape="circle",
    fill="red!60",
    draw="red",
)

fig.show()
print(f"Centroid: ({centroid.x}, {centroid.y})")
print(f"Expected: ({(0 + 6 + 2) / 3}, {(0 + 0 + 4) / 3})")
```
