# Tutorial 5 - 3D plotting

This tutorial adds raw tikz code in a TikzFigure.

```python
%load_ext autoreload
%autoreload 2
```

```python
from tikzpics import TikzFigure
```

```python
fig = TikzFigure()

n1 = fig.add_node(0, 0, shape="circle", color="white", fill="blue", content="Hello!")
n2 = fig.add_node(5, 0, shape="circle", color="white", fill="red", content="Hi!")

fig.add_raw("% This is a comment in the TikZ code")
fig.add_raw("\\node at (2.5, 1) {This is a raw TikZ node};")
fig.add_raw(f"\\draw[->, thick, dashed, in=30] ({n2.label}) -- ({n1.label});")

fig.show()
print(fig.generate_tikz())
```
